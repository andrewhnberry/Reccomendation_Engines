# One Hot Encode Genres

```python
df_combined['genres'][4]
```

This looks like json Data

```python
genres_json = df_combined['genres'].apply(json.loads)
genres_json
```

```python
genres_df = pd.DataFrame(
{
    'id' : df_combined['id'].values.repeat(genres_json.str.len(), axis=0),
    'genre' : np.concatenate(genres_json.tolist())
})
```

```python
genres_df['genre'] = genres_df['genre'].map(lambda x: x.get('name'))
```

```python
genres_df
```

```python
#Use pandas get_dummies function
dummies_genres = genres_df.set_index('id').genre.str.get_dummies().sum(level=0)
```

```python
#dummies_genres['id_movie'] = dummies_genres.index
dummies_genres = dummies_genres.reset_index()
```

```python
dummies_genres.head(2)
```

```python
#Merge the two dataframes
df_3 = df_combined.merge(dummies_genres, on='id')
```

```python
df_3 = df_3.drop(columns = ['genres'])
```

### One Hot Encode Keywords

```python
keywords_json = df_3['keywords'].apply(json.loads)
```

```python
keywords_json[0]
```

```python
keywords_df = pd.DataFrame(
{
    'id' : df_3['id'].values.repeat(keywords_json.str.len(), axis=0),
    'keywords' : np.concatenate(keywords_json.tolist())
})

keywords_df['keywords'] = keywords_df['keywords'].map(lambda x: x.get('name'))
```

```python
#Use pandas get_dummies function
dummies_keywords = keywords_df.set_index('id').keywords.str.get_dummies().sum(level=0)

#Reset Index
dummies_keywords = dummies_keywords.reset_index()
```

```python
dummies_keywords.shape
```

```python
#Merge the two dataframes
df_4 = df_3.merge(dummies_keywords, on='id')
```

#### tfidf Vectorize Overview

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words = 'english')
```

```python
df_4['overview'] = df_4['overview'].fillna('')
```

```python
tfidf_matrix = tfidf.fit_transform(df_4.overview)
```

```python
tfidf_matrix.shape
```

### Cosine similiarity Score

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(df_4.index, index=df_4['title']).drop_duplicates()
```

```python
#Cosine FUnction

def movie_reccomendation(movie, cosine_sim = cosine_sim):
    
    idx = indices[movie]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    
    sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse = True)
    
    
    # Top 10 movies based on similar-ness
    sim_scores = sim_scores[1:11]
    
    movie_indices = [i[0] for i in sim_scores]
    
    
    return df_4.title.iloc[movie_indices]
```
